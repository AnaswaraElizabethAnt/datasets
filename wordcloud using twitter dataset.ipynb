```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re
```

```python
tweets=pd.read_csv("D:/dataset/BJP4Karnataka_tweets.csv")
tweets.head()
```

```python
def wc(tweets,col):
    words = ' '.join(tweets[col])
    colmask = np.array(Image.open("D:/dataset/Karnataka.png"))
    stopwords= set(STOPWORDS)
    word_mod=re.sub("[^a-zA-Z0-9_]+"," ",words)
    word_mod=' '.join([word for word in word_mod.split()
                      if "http" not in word
                      and word !='RT'
                      and word !='co'])
    bjp_wc = WordCloud(background_color='black',mask=colmask,stopwords = stopwords,min_font_size=16).generate(word_mod)
    plt.figure(figsize=(20,10))
    plt.imshow(bjp_wc)
    plt.tight_layout(pad=0)
    plt.axis("off")
    plt.show()
```

```python
wc(tweets,'text')
```
